## Visualization


To visualize the distribution of the features and the labels, we transform the shares in binary values


--> we suggest the team 2 that they first remove outliers and noises and then decide whether they want to use the median or mean for the split, but now we use the median because the descriptive statistics showed that the mean and median have a large difference (keep in mind that the median is less affected by outliers)


```python
median = pop.median()[' shares']
display(median)

threshold = median
pop.loc[pop[' shares'] > threshold, 'will be popular'] = 1
pop.loc[pop[' shares'] <= threshold, 'will be popular'] = 0
display(pop)
```

## Feature Distribution

```python
row = 12
column = 4
data = pop.drop('will be popular', axis=1).drop(' shares', axis=1)
title = "Feature distributions"

fig = plt.figure(figsize = (45, 45))
fig.suptitle(title, fontsize=25, y = 0.9)

features_length = data.shape[1]
for i in range(features_length):
    ax = fig.add_subplot(row, column, i + 1)
    sns.distplot(data.iloc[:, i], kde=False)
plt.show()
```

--> The features have different distributions

--> some have an almost normal distribution like sentiment_subjectivity

--> binary features are distributed at value 0 or 1, more of the features have the value 0, this makes sense because only one day and category is selected per row

--> some features have a sknewness (for example feature max_polarity_in_negative_words : this makes sense because the polarity is a value lower than 0 where -1 is the maximum of negativesness)

```python
from datasets import wine_red_dataset, wine_white_dataset
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
wine_red = wine_red_dataset()
wine_white = wine_white_dataset()
```

```python
# DataFrame fur Einheiten aus der realen Welt
import pandas as pd
feature_units = pd.DataFrame([
    ['mg/l'],
    ['g/l'],
    #['density', 'g/l']
    ], columns=['unit'], index=['flavanoids', 'citric acid'])
display(feature_units)

# utility function to plot from our dataset
def show_scatterplot(feature, label, feature_label, label_label):
    plt.scatter(y=wine_red[feature], x=wine_red[label])
    plt.xlabel(label_label)
    plt.ylabel(feature_label)
    plt.show()
```

### Flavanoide
**Flavanoide** (wasserlosliche Pflanzenfarbstoffe) ist verantwortlich fur den Geschmack, den Geruch und die Farbe des Weines

Es ist erkennbar, dass im Datensatz Flavanoide sehr ahnliche Werte aufweisen und die Qualitat bei diesen sehr verteilt ist

## Qualitat in categorical values (gut/schlecht)

```python
plt.hist(wine_red['quality'])
plt.show()
```

```python
quality_threshold = 5
wine_red.loc[wine_red['quality'] > quality_threshold, 'quality_cat'] = 1
wine_red.loc[wine_red['quality'] <= quality_threshold, 'quality_cat'] = 2
display(wine_red)
```

```python
plt.hist(wine_red['quality_cat'])
plt.show()
```

```python
wine_red.corrwith(wine_red['quality']).plot.bar(
        figsize = (15, 7.5), title = "Correlation with quality", fontsize = 15,
        rot = 45, grid = True)
```

- the quality seems to raise with raising alcohol
- the quality seems to decrease when the lightness also decreases
- the quality seems to decrease when the volatile acidity also decreases

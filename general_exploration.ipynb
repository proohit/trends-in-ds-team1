```python
from datasets import wine_red_dataset, wine_white_dataset
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
wine_red = wine_red_dataset()
wine_white = wine_white_dataset()
```

Descriptive statistics
